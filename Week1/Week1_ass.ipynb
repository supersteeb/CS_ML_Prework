```python
import pandas as pd
import numpy as np
import csv
```

```python
songs_dataset = pd.read_json('MasterSongList.json')
```

```python
songs_dataset.head(10)
```

```python
songs_dataset.loc[:, 'genres'] = songs_dataset['genres'].apply(''.join)

def consolidateGenre(genre):
    if (len(genre) > 0):
        return genre.split(':')[0]
    else:
        return genre
    
songs_dataset.loc[:, 'genres'] = songs_dataset['genres'].apply(consolidateGenre)
```

```python
songs_dataset.head(10)
```

```python
audioFeatureList = []
for audio_feature in songs_dataset['audio_features']:
    audioFeatureList.append(audio_feature)
```

```python
audio_feature_headers = ['key', 'energy', 'liveliness', 'tempo', 'speechiness', 'acousticness', 'instrumentalness', 'time_signature', 'duration', 'loudness', 'valence', 'danceability', 'mode', 'time_signature_confidence', 'tempo_confidence', 'key_confidence', 'mode_confidence']

len(audio_feature_headers)
```

```python
df_features = pd.DataFrame(audioFeatureList, columns=audio_feature_headers)
df_features.head(10)
```

```python
df_features['genres'] = songs_dataset['genres']
```

```python
df_features[df_features['genres'] == 'country']
```

```python
consolidatedRapCountry = songs_dataset.copy()
```

```python
consolidatedRapCountry = consolidatedRapCountry[(songs_dataset['genres'] == 'rap') | (songs_dataset['genres'] == 'country')]
```

```python
consolidatedRapCountry.reset_index(drop=True, inplace=True)
consolidatedRapCountry
```

```python
audioFeatureList2 = []
for audio_feature in consolidatedRapCountry['audio_features']:
    audioFeatureList2.append(audio_feature)

df_features2 = pd.DataFrame(audioFeatureList2, columns=audio_feature_headers)
```

```python
df_features2 = df_features2.dropna()
df_features2.head()
```

```python
df_features2['genres'] = consolidatedRapCountry['genres']
rap_df_features2 = df_features2[df_features2['genres'] == 'rap'].sample(n=1075, replace=True)
rap_df_features2
#1075 country <= lower of the 2
```

# Balancing the Data
### there are 1075 country  

```python
cRC_features = pd.concat([df_features2[df_features2['genres'] == 'country'], rap_df_features2])
cRC_features2 = cRC_features.copy()
cRC_features = cRC_features.drop('genres', 1)
cRC_features
```

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
min_max_scaler = MinMaxScaler()
```

```python
cRC_scaled_features = min_max_scaler.fit_transform(cRC_features)
cRC_scaled_features
```

# KNN Classifier

```python
np.round(cRC_scaled_features, decimals=3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(cRC_scaled_features, cRC_features2['genres'], test_size=0.3, random_state=101)
```

```python
knn.fit(X_train, y_train)
```

```python
predictions = knn.predict(X_test)
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```

```python
from nltk import ConfusionMatrix
print(ConfusionMatrix(list(y_test), list(predictions)))
```

```python
classification_report(y_test, predictions)
```

# Cross Validation

```python
from sklearn.cross_validation import cross_val_score
```

```python
scores = cross_val_score(knn, cRC_scaled_features, cRC_features2['genres'], cv=10)
```
